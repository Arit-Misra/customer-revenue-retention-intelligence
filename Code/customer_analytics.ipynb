```python
import pandas as pd

BASE_PATH = r"C:\Users\Arit\Desktop\IITB\Self Project\Customer Revenue & Retention Intelligence Platform\archive"
```

```python
orders = pd.read_csv(f"{BASE_PATH}\olist_orders_dataset.csv")
customers = pd.read_csv(f"{BASE_PATH}\olist_customers_dataset.csv")
items = pd.read_csv(f"{BASE_PATH}\olist_order_items_dataset.csv")
payments = pd.read_csv(f"{BASE_PATH}\olist_order_payments_dataset.csv")
products = pd.read_csv(f"{BASE_PATH}\olist_products_dataset.csv")
reviews = pd.read_csv(f"{BASE_PATH}\olist_order_reviews_dataset.csv")
```

```python
datasets = {
    "orders": orders,
    "customers": customers,
    "items": items,
    "payments": payments,
    "products": products,
    "reviews": reviews
}

for name, df in datasets.items():
    print(f"\n{name.upper()}")
    print("Shape:", df.shape)
    print("Columns:", df.columns.tolist())
    display(df.head(3))
```

```python
for name, df in datasets.items():
    print(f"\n{name.upper()} Missing Values (%)")
    display((df.isnull().mean() * 100).round(2))
```

```python
print("Orders unique:", orders["order_id"].nunique())
print("Items unique:", items["order_id"].nunique())
```

```python
date_cols = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_cols:
    orders[col] = pd.to_datetime(orders[col], errors="coerce")
```

```python
order_value = (
    items.groupby("order_id", as_index=False)["price"]
    .sum()
    .rename(columns={"price": "order_value"})
)
```

```python
payments_clean = (
    payments.groupby("order_id", as_index=False)
    .agg({
        "payment_type": "first",
        "payment_value": "sum"
    })
)
```

```python
df = (
    orders
    .merge(customers, on="customer_id", how="left")
    .merge(order_value, on="order_id", how="left")
    .merge(payments_clean, on="order_id", how="left")
    .merge(reviews[["order_id", "review_score"]], on="order_id", how="left")
)
```

```python
print("Rows:", df.shape[0])
df[["order_value", "review_score"]].describe()
df.isnull().mean().sort_values(ascending=False).head(8)
```

```python
df["order_value"] = df["order_value"].fillna(0)
df["review_score"] = df["review_score"].fillna(df["review_score"].median())
```

```python
df["delivery_days"] = (
    df["order_delivered_customer_date"] -
    df["order_purchase_timestamp"]
).dt.days
```

```python
# Order month
df["order_month"] = df["order_purchase_timestamp"].dt.to_period("M").astype(str)

# First purchase date per customer
first_order = (
    df.groupby("customer_unique_id")["order_purchase_timestamp"]
      .min()
      .reset_index()
      .rename(columns={"order_purchase_timestamp": "first_order_date"})
)

df = df.merge(first_order, on="customer_unique_id", how="left")

# Cohort month
df["cohort_month"] = pd.to_datetime(df["first_order_date"]).dt.to_period("M").astype(str)
```

```python
order_counts = (
    df.groupby("customer_unique_id")["order_id"]
      .transform("nunique")
)

df["is_repeat_customer"] = (order_counts > 1).astype(int)
```

```python
snapshot_date = df["order_purchase_timestamp"].max()

rfm = (
    df.groupby("customer_unique_id")
      .agg(
          recency=("order_purchase_timestamp", lambda x: (snapshot_date - x.max()).days),
          frequency=("order_id", "nunique"),
          monetary=("order_value", "sum")
      )
      .reset_index()
)
```

```python
rfm["R_score"] = pd.qcut(rfm["recency"], 4, labels=[4,3,2,1]).astype(int)
rfm["F_score"] = pd.qcut(rfm["frequency"].rank(method="first"), 4, labels=[1,2,3,4]).astype(int)
rfm["M_score"] = pd.qcut(rfm["monetary"], 4, labels=[1,2,3,4]).astype(int)

rfm["RFM_score"] = rfm["R_score"] + rfm["F_score"] + rfm["M_score"]
```

```python
rfm.describe()
rfm["RFM_score"].value_counts().sort_index()
df[["order_month", "cohort_month", "is_repeat_customer"]].head()
```

```python
monthly_revenue = (
    df.groupby("order_month", as_index=False)
      .agg(
          revenue=("order_value", "sum"),
          orders=("order_id", "nunique"),
          customers=("customer_unique_id", "nunique")
      )
)

monthly_revenue["aov"] = monthly_revenue["revenue"] / monthly_revenue["orders"]
monthly_revenue["mom_growth_pct"] = monthly_revenue["revenue"].pct_change() * 100
```

```python
cohort_counts = (
    df.groupby(["cohort_month", "order_month"])["customer_unique_id"]
      .nunique()
      .reset_index(name="active_customers")
)

cohort_sizes = (
    cohort_counts.groupby("cohort_month")["active_customers"]
                 .first()
                 .reset_index(name="cohort_size")
)

cohort_retention = cohort_counts.merge(cohort_sizes, on="cohort_month")
cohort_retention["retention_rate"] = (
    cohort_retention["active_customers"] / cohort_retention["cohort_size"]
)
```

```python
df = df.merge(
    rfm[["customer_unique_id", "RFM_score"]],
    on="customer_unique_id",
    how="left"
)
```

```python
def segment_label(score):
    if score >= 10:
        return "High Value"
    elif score >= 7:
        return "Mid Value"
    else:
        return "Low Value"

rfm["segment"] = rfm["RFM_score"].apply(segment_label)
```

```python
ops_metrics = (
    df.groupby("order_month", as_index=False)
      .agg(
          avg_delivery_days=("delivery_days", "mean"),
          avg_review_score=("review_score", "mean"),
          late_delivery_rate=("delivery_days", lambda x: (x > 10).mean())
      )
)
```
